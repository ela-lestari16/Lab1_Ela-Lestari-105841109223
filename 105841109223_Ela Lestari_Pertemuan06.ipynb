```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score

# ============================================================
# Load Dataset
# ============================================================
data = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42
)

# ============================================================
# Tugas 1: Bandingkan DT vs RF vs GB (Classification)
# ============================================================
print("=" * 55)
print("Tugas 1: Perbandingan DT vs RF vs GB (Classification)")
print("=" * 55)

models = {
    "DecisionTree": DecisionTreeClassifier(random_state=42),
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=42),
    "GradientBoosting": GradientBoostingClassifier(random_state=42),
}

results = {}

for name, model in models.items():

    model.fit(X_train, y_train)

    pred = model.predict(X_test)

    acc = round(accuracy_score(y_test, pred), 3)

    results[name] = acc

    print(f"{name:20s} accuracy: {acc}")
```

```python
# ============================================================
# Tugas 2: Top 10 Feature Importance dari Random Forest
# ============================================================
print("\n" + "=" * 55)
print("Tugas 2: Top 10 Feature Importance (Random Forest)")
print("=" * 55)

rf = models["RandomForest"]

importances = rf.feature_importances_

indices = np.argsort(importances)[::-1][:10]

for rank, idx in enumerate(indices, 1):

    print(f"{rank:2d}. {data.feature_names[idx]:30s} importance: {importances[idx]:.4f}")
```

```python
# ============================================================
# Tugas 3: Ubah n_estimators dan lihat pengaruhnya
# ============================================================
print("\n" + "=" * 55)
print("Tugas 3: Pengaruh n_estimators pada Random Forest")
print("=" * 55)

best_acc = 0
best_n = 0

for n in [10, 50, 100, 200, 500]:

    rf_n = RandomForestClassifier(n_estimators=n, random_state=42)

    rf_n.fit(X_train, y_train)

    acc = accuracy_score(y_test, rf_n.predict(X_test))

    print(f"n_estimators={n:>3d} -> accuracy: {acc:.3f}")

    # menyimpan model terbaik
    if acc > best_acc:
        best_acc = acc
        best_n = n

# ============================================================
# Model Terbaik
# ============================================================
print("\n" + "=" * 55)
print("Model Terbaik Random Forest")
print("=" * 55)

print(f"n_estimators terbaik = {best_n}")
print(f"Akurasi terbaik     = {best_acc:.3f}")

# ============================================================
# Penjelasan Pengaruh n_estimators
# ============================================================
print("\nAlasan:")
print("1. n_estimators adalah jumlah pohon keputusan dalam Random Forest.")
print("2. Semakin banyak pohon, model biasanya menjadi lebih stabil karena prediksi menggunakan voting dari banyak tree.")
print("3. Namun jika jumlah pohon terlalu banyak, peningkatan akurasi tidak selalu signifikan dan waktu komputasi menjadi lebih lama.")
```
