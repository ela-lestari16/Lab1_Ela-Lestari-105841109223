```python
# import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Membuat array random 20 angka
np.random.seed(10)
arr = np.random.randint(1, 100, 20)

print("Array:", arr)
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))
```

```python
# Membuat DataFrame nilai 10 mahasiswa
data = {
    "Nama": ["Ani","Rara","Dido","Dina","Eka",
             "Farah","Gina","Caca","Intan","Joko"],
    "Nilai": np.random.randint(60, 100, 10)
}

df = pd.DataFrame(data)

print("\nData Nilai Mahasiswa:")
print(df)

# 3 nilai tertinggi
top3 = df.sort_values(by="Nilai", ascending=False).head(3)
print("\n3 Nilai Tertinggi:")
print(top3)
```

```python
# Grafik 1: Bar Chart
df.plot(kind="bar", x="Nama", y="Nilai", legend=False)
plt.title("Grafik Nilai Mahasiswa")
plt.ylabel("Nilai")
plt.show()

# Grafik 2: Histogram
plt.hist(df["Nilai"], bins=5)
plt.title("Distribusi Nilai Mahasiswa")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.show()
```
